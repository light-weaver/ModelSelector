```python
from modelselector.CalculateFeatures import calculatefeatures
from modelselector.generateData import create_datasets

import glob
import pandas as pd
```

```python
# To create datasets
create_datasets()
```

```python
root_folder = "./datasets_benchmark/"
all_files = glob.glob(root_folder + "/*.csv")
```

```python
features = pd.DataFrame()
for file in all_files:
    data = pd.read_csv(file)
    input_columns = [column for column in data.columns if 'x' in column]
    inputs = data[input_columns].to_numpy()
    output_columns = [column for column in data.columns if 'f' in column]
    output = output_columns[-1]
    output = data[output].to_numpy()
    features = features.append(calculatefeatures(inputs, output))
# Cleaning the file names
all_files[:] = [file.split('/')[-1] for file in all_files]
pd.Series(all_files, name='filename')
```
