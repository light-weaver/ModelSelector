```python
class DTLZ6(dtlz.DTLZ5):
    """DTLZ6 only differs from DTLZ5 in the g function."""

    def g(self, phenome):
        """The g function of DTLZ6."""
        g = 0.0
        n = self.num_variables
        k = n - self.num_objectives + 1
        for i in range(n - k + 1, n + 1):
            g += pow(phenome[i-1] -0.5, 0.1)
        return g
```
